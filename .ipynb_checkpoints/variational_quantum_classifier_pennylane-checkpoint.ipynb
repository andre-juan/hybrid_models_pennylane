```python
import pandas as pd

from sklearn.model_selection import train_test_split

import tensorflow as tf

import pennylane as qml
from pennylane import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import random
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

def clf_metrics(estimator, X_test, y_test, train=False, X_train=None, y_train=None):
    '''
    função para visualização de métricas de classificação: matriz de confusão 
    e classification report.
    opcionalmente, exibimos também as métricas para os dados de treino
    '''
    
    if train:
        
        y_prob_train = estimator.predict(X_train)
        y_pred_train = y_prob_train.argmax(axis=-1)

        print("\nMétricas de classificação nos dados de treino:")

        ConfusionMatrixDisplay.from_predictions(y_train, y_pred_train)
        plt.show()

        print(classification_report(y_train, y_pred_train))
    
    # ================================================
    
    y_prob_test = estimator.predict(X_test)
    y_pred_test = y_prob_test.argmax(axis=-1)
    
    print("\nMétricas de classificação nos dados de teste:")
    
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred_test)
    plt.show()
    
    print(classification_report(y_test, y_pred_test))
```

```python
def plot_history(history):
    '''
    função para plotar a evolução da função de perda e acurácia em função dos epochs.
    informações plotadas para dados de treino e validação.
    '''
    
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(len(loss))

    plt.figure(figsize=(10, 5))
    plt.plot(epochs, loss, 'red', label = 'Loss de treino')
    plt.plot(epochs, val_loss, 'blue', label = 'Loss de validação')
    plt.title('Loss de treino e validação')
    plt.legend()
    plt.show()
    
    # ================================================

    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']

    epochs = range(len(loss))

    plt.figure(figsize=(10, 5))
    plt.plot(epochs, acc, 'red', label = 'Acurácia de treino')
    plt.plot(epochs, val_acc, 'blue', label = 'Acurácia de validação')
    plt.title('Acurácia de treino e validação')
    plt.legend()
    plt.show()
```

```python
def set_seeds(seed=42):
    '''
    função que define as seeds para reproducibilidade dos resultados
    '''
    
    np.random.seed(seed)
    tf.random.set_seed(seed)
    random.seed(seed)
    tf.keras.utils.set_random_seed(seed)
    tf.config.experimental.enable_op_determinism()
```

Vamos agora conhecer o **Classificador Variacional Quântico** (conhecido como VQC, do inglês **Variational Quantum Classifier**)

### VQC: algoritmo em camadas

O circuito quântico do VQC pode ser esquematicamente visualizado em 3 componentes, ou camadas:

- **Feature map $\mathcal{U}_{\Phi(\vec{x})}$**: parte do circuito responsável por codificar os dados clássicos $\vec{x}$ em estados quânticos que serão processados no circuito quântico pelo algoritmo. Ou seja, é nesse passo que a informação clássica é representada na forma de informação quântica;<br><br>

- **Variational layer $W(\vec{\theta})$**: parte do circuito parametrizada por $\vec{\theta}$. Esses são os parâmetros ajustados no processo de treinamento;<br><br>

- **Measurement**: último passo do circuito, que consiste em medidas dos qubits, o que produz informação clássica.<br><br>

<img src=https://www.researchgate.net/profile/Claudio-Gambella-2/publication/342464434/figure/fig4/AS:906517855358979@1593142107809/Quantum-circuit-for-Variational-Quantum-Classifier-VQC-that-consists-of-fixed-quantum.jpg width=50%>

Na figura acima, o feature map e a camada variacional foram representados esquematicamente por "blocos", que representam composições de gates quânticos, que podem ser construídas das mais diversas formas.

A escolha particular para a sequência de gates quânticos específica a ser utilizada é conhecida como a **arquitetura** do circuito variacional, e há enorme liberdade quanto à arquitetura, como veremos a seguir.

### Um método híbrido

O VQC é um **algoritmo híbrido**, isto é, a rotina completa é **parte clássica** e **parte quântica**.

Podemos imaginar desta maneira:

- O "**forward propagation**" é quântico;<br><br>

- Com a medida, recuperamos informação clássica, e construímos a função de perda/custo classicamente;<br><br>

- Os parâmetros $\vec{\theta}$ são atualizados de modo a otimizar classicamente a função de custo;<br><br>

- Neste sentido, o "**backpropagation**" é clássico.

<img src=https://media.springernature.com/original/springer-static/image/chp%3A10.1007%2F978-3-030-50433-5_45/MediaObjects/500809_1_En_45_Fig1_HTML.png width=50%>

Para além disso, a seguir construiremos arquiteturas em que o forward propagation também será híbrido (isto é, uma rede neural com camadas clássicas e camadas quânticas, em uma única estrutura sequencial).

### Variacional?!

O "variacional" que nomeia o VQC tem origem nos [métodos variacionais](https://arxiv.org/abs/1812.08767) da mecânica quântica.

Historicamente, estes princípios variacionais têm sido utilizados como métodos aproximativos, notoriamente em problemas de [simulação quântica](https://arxiv.org/abs/1308.6253).

Alguns algoritmos quânticos foram propostos com base nestes métodos, em particular:

- [VQE](https://arxiv.org/abs/2111.05176), que foi introduzido no contexto de simulação quântica;

- [QAOA](https://arxiv.org/abs/1411.4028), que foi introduzido para endereçar problemas de otimização combinatória.

Ambos estes algoritmos são **rotinas híbridas**. Por este motivo, cunhou-se o termo **"algoritmos variacionais"** para designar esta abordagem híbrida, que envolve **circuitos quânticos parametrizados** cujos parâmetros são otimizados classicamente.

## Leitura de dados - good old iris


```python
df = pd.read_csv("iris.csv")

X = df.drop(columns="species")
y = df["species"]

# ================================================

# dados de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    stratify=y, random_state=42)
# dados de validação pra rede neural
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, 
                                                  stratify=y_train, random_state=42)

# ================================================

# classes representadas numericamente
y_train_number = y_train.astype("category").cat.codes
y_val_number = y_val.astype("category").cat.codes
y_test_number = y_test.astype("category").cat.codes

# classes representadas como vetores binários, importante pra rede neural
y_train_tf = tf.one_hot(y_train_number, depth=3)
y_val_tf = tf.one_hot(y_val_number, depth=3)
```

```python
sns.pairplot(df, hue="species");
```

## Arquitetura do circuito variacional


```python
# usaremos 1 qubit pra cada feature
n_qubits = X_train.shape[1]
```

```python
# esses são os backends de execução dos circuitos quânticos da pennylane
# usaremos apenas simuladores

# esse é o backend padrão que usaremos
dev = qml.device("default.qubit", wires=n_qubits)

# esse backend usaremos apenas pra visualizar os circuitos, usando as funcionalidades do qiskit
dev_qiskit = qml.device("qiskit.aer", wires=n_qubits, seed_simulator=42)
```

### Feature map


```python
X_train.iloc[0]
```

```python
@qml.qnode(dev_qiskit)
def feature_map(inputs):
    
    qml.AngleEmbedding(inputs, wires=range(n_qubits))
    
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

# aqui estamos passando a primeira observação de treino para ser carregada no circuito quântico
# o valor de cada feature será o ângulo de um gate de rotação (por isso o nome "angle embedding")
print(feature_map(X_train.iloc[0].values))

dev_qiskit._circuit.draw("mpl")
```

Uma outra possibilidade de feature map (ZZ feature map), introduzida pela IBM [neste artigo:](https://arxiv.org/pdf/1804.11326.pdf)

```python
@qml.qnode(dev_qiskit)
def ZZ_feature_map(inputs):
    
    qml.IQPEmbedding(inputs, wires=range(n_qubits))
    
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

print(ZZ_feature_map(X_train.iloc[0].values))

dev_qiskit._circuit.draw("mpl")
```

### Variational layer

```python
@qml.qnode(dev_qiskit)
def var_layer(parametros):
    
    qml.StronglyEntanglingLayers(parametros, wires=range(n_qubits))
    
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

# 1 camada variacional 
# (primeiro pegamos o shape, que dá a quantidade de parâmetros,
# e aí geramos parametros iniciais aleatórios)
shape = qml.StronglyEntanglingLayers.shape(n_layers=1, n_wires=4)
np.random.seed(42)
parametros = np.random.random(size=shape)

print(var_layer(parametros))

dev_qiskit._circuit.draw("mpl")
```

Mais uma camada:

```python
@qml.qnode(dev_qiskit)
def var_layer(parametros):
    
    qml.StronglyEntanglingLayers(parametros, wires=range(n_qubits))
    
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

# 2 camadas variacionais agora
shape = qml.StronglyEntanglingLayers.shape(n_layers=2, n_wires=4)
np.random.seed(42)
parametros = np.random.random(size=shape)

print(var_layer(parametros))

dev_qiskit._circuit.draw("mpl")
```

### Juntando tudo

Essa é a arquitetura do circuito variacional que utilizaremos: angle embedding para o feature map, e duas camadas variacionais do tipo strong entangling

```python
@qml.qnode(dev_qiskit)
def vqc_layer(inputs, parametros):
    
    qml.AngleEmbedding(inputs, wires=range(n_qubits))
    
    qml.StronglyEntanglingLayers(parametros, wires=range(n_qubits))
    
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

shape = qml.StronglyEntanglingLayers.shape(n_layers=2, n_wires=4)
np.random.seed(42)
parametros = np.random.random(size=shape)

print(vqc_layer(X_train.iloc[0].values, parametros))

dev_qiskit._circuit.draw("mpl")
```

### Agora vamos treinar os modelos híbridos!

___________________

## Modelo híbrido: forward propagation quântico, backpropagation clássico

Neste caso, usamos apenas uma camada quântica no forward propagation:

<img src="figuras/vqc_quantum_only.PNG" width=800>

```python
set_seeds(42)

n_qubits = X_train.shape[1]
n_classes = y_train.nunique()
n_var_layers = 2

# esse é um parâmetro para a construção da camada keras a partir do nó quântico
# o "3" no final é fixo, e representa os 3 ângulos necessários para cada bloco de rotações
# da arquitetura variacional "strongly entangling"
weight_shapes = {"parametros" : (n_var_layers, n_qubits, 3)}

# note que agora estamos usando o dev padrão!
@qml.qnode(dev)
def vqc_layer(inputs, parametros):
    
    qml.AngleEmbedding(inputs, wires=range(n_qubits))
    
    qml.StronglyEntanglingLayers(parametros, wires=range(n_qubits))
    
    # aqui, vamos medir apenas os 3 (n_classes) primeiros qubits, um pra cada classe!
    return [qml.expval(qml.PauliZ(i)) for i in range(n_classes)]

# ================================================

model = tf.keras.models.Sequential()

# é aqui que transformamos o qnode (circuito variacional quântico) em uma camada do keras
# para que seja possível otimizar os parâmetros normalmente através do backpropagation,
# com as ferramentas do keras (classicamente).
# note que estamos utilizando a pennylane (importada como qml) para esse fim!!
model.add(qml.qnn.KerasLayer(vqc_layer, weight_shapes, output_dim=n_classes))

model.add(tf.keras.layers.Activation("softmax"))

# ================================================
# otimização clássica

opt = tf.keras.optimizers.Adam(learning_rate=0.05)

model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])

history = model.fit(X_train, y_train_tf,
                    epochs=10, batch_size=10,
                    validation_data=(X_val, y_val_tf))

# ================================================
```

```python
plot_history(history)
clf_metrics(model, X_test, y_test_number, True, X_train, y_train_number)
```

## Modelo híbrido: camada densa clássica antes do output

Aqui, adicionamos uma camada clássica após a camada quântica. Com isso, o modelo é híbrido até mesmo no forward propagation:

<img src="figuras/vqc_classical_before_output.PNG" width=800>

```python
set_seeds(42)

n_qubits = X_train.shape[1]
n_classes = y_train.nunique()
n_var_layers = 2

weight_shapes = {"parametros" : (n_var_layers, n_qubits, 3)}

@qml.qnode(dev)
def vqc_layer(inputs, parametros):
    
    qml.AngleEmbedding(inputs, wires=range(n_qubits))
    
    qml.StronglyEntanglingLayers(parametros, wires=range(n_qubits))
    
    # agora, medimos todos os qubits! os valores medidos serão passados
    # como input para a camada clássica seguinte!
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

model = tf.keras.models.Sequential()

model.add(qml.qnn.KerasLayer(vqc_layer, weight_shapes, output_dim=n_qubits))

model.add(tf.keras.layers.Dense(n_classes, activation="softmax"))

# ================================================

opt = tf.keras.optimizers.Adam(learning_rate=0.05)

model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])

history = model.fit(X_train, y_train_tf,
                    epochs=10, batch_size=10,
                    validation_data=(X_val, y_val_tf))

# ================================================
```

```python
plot_history(history)
clf_metrics(model, X_test, y_test_number, True, X_train, y_train_number)
```

## Modelo híbrido: mais uma camada densa clássica

Mais um modelo com forward propagation híbrido. Aqui, adicionamos uma camada clássica também antes da camada quântica:

<img src="figuras/vqc_two_classical.PNG" width=800>

```python
set_seeds(42)

n_qubits = X_train.shape[1]
n_classes = y_train.nunique()
n_var_layers = 2

weight_shapes = {"parametros" : (n_var_layers, n_qubits, 3)}

@qml.qnode(dev)
def vqc_layer(inputs, parametros):
    
    qml.AngleEmbedding(inputs, wires=range(n_qubits))
    
    qml.StronglyEntanglingLayers(parametros, wires=range(n_qubits))
    
    # agora, medimos todos os qubits! os valores medidos serão passados
    # como input para a camada clássica seguinte!
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# ================================================

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(n_qubits, activation="relu", input_dim=X_train.shape[1]))

model.add(qml.qnn.KerasLayer(vqc_layer, weight_shapes, output_dim=n_qubits))

model.add(tf.keras.layers.Dense(n_classes, activation="softmax"))

# ================================================

opt = tf.keras.optimizers.Adam(learning_rate=0.05)

model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])

history = model.fit(X_train, y_train_tf,
                    epochs=10, batch_size=10,
                    validation_data=(X_val, y_val_tf))

# ================================================
```

```python
plot_history(history)
clf_metrics(model, X_test, y_test_number, True, X_train, y_train_number)
```
